```python
def convert_to_people(cell):
    if cell == "n.a.":
        return "Sami Shiraz"
    return cell
def convert_to_peoplee(cell):
    if cell == "not available":
        return None
    return cell
df = pd.read_excel('stock_data.xlsx', converters={
    'people' : convert_to_people,
    'eps' : convert_to_peoplee
})
```

```python
df.to_excel("sami.xlsx",sheet_name='Stock', index=False, startcol=2,startrow=2)
```
