```python
import pandas as pd
```

```python
daa = pd.read_csv('wheater_data.csv', parse_dates=['day'])
daa.set_index('day', inplace=True)
```

```python
daa.fillna(0)
```

```python
new_daa = daa.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'No Event'
})
new_daa
```

```python
new_daa = daa.fillna(method='ffill')
```

```python
new_daa = daa.fillna(method='bfill', limit=1)
new_daa
```

```python
new_daa = daa.fillna(method='bfill', axis='columns')
new_daa
```

```python
new_daa = daa.interpolate()
new_daa
```

```python
new_daa = daa.interpolate(method='time')
new_daa
```

```python
new_daa = daa.dropna()
new_daa
```

```python
new_daa = daa.dropna(how='all')
new_daa
```

```python
new_daa = daa.dropna(thresh=1)
new_daa
```

```python
new_daa = daa.dropna(thresh=2)
new_daa
```
